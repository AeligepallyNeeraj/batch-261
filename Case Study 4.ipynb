# 7.For each month - how many Data Bank customers make more than 1 deposit and either 1 purchase or 1 withdrawal in a single month?

```python
C_T
```

```python
import pandas as pd

# Sample DataFrame with a date column
data = C_T['txn_date']
df=pd.DataFrame(data)

# Convert the 'date' column to datetime
df['date'] = pd.to_datetime(data)

# Extract the month from the 'date' column and create a new 'month' column
df['month'] = df['date'].dt.month

```

```python
C_T['txn_date'] =(C_T['txn_date']).dt.month
```

```python
deposits = C_T[C_T['txn_type'] == 'deposit']
```

```python
customers_with_purchase_or_withdrawal = C_T[C_T['txn_type'].isin(['purchase', 'withdrawal'])]['customer_id'].unique()
```

```python
monthly_activity = deposits[deposits['customer_id'].isin(customers_with_purchase_or_withdrawal)]
```

```python
monthly_activity = monthly_activity.groupby(monthly_activity['txn_date']).agg({'customer_id': 'nunique'}).reset_index()
```

```python
monthly_activity.columns = ['month', 'customers_count']
```

```python
monthly_activity
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# Plotting using seaborn barplot
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='month', y='customers_count', data=monthly_activity, ci=None)

# Add legend
ax.legend(labels=['Number of Customers'], loc='upper left')

plt.title('Monthly Customer Activity')
plt.xlabel('Month')
plt.ylabel('Number of Customers')
plt.show()
```

# The 1st month has the highest no of  customers count
# The 4th month has the lowest no of customers count
