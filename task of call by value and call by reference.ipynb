# call by value

* In call by value method of parameter passing, the values of actual parameters are copied to the functionâ€™s formal parameters.

* There are two copies of parameters stored in different memory locations.
* One is the original copy and the other is the function copy.
* Any changes made inside functions are not reflected in the actual parameters of the caller.

```python
def increment(a):
    a+=1
num=1
increment(num)
num
```

# call by refernce

* In Python, arguments are passed by object reference. However, it's essential to clarify that Python does not have true "call by reference" like some other languages. Instead, it uses a mechanism often referred to as "call by object reference" or "call by sharing."

* In Python, when you pass a mutable object (like a list or dictionary) to a function, changes made to the object inside the function are reflected outside the function as well. This is because the reference to the object is passed, not a copy of the object.`

```python
def modify_list(my_list):
    my_list.append(4)
    my_list[0] = 100

my_numbers = [1, 2, 3]
modify_list(my_numbers)

print(my_numbers)
```
