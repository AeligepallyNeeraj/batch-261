```python
tip=x[:,2 ]#avg tip of male and female
sex=x[:,3]
total_tip_male = np.mean(tip[sex == 0])
total_tip_female = np.mean(tip[sex == 1])
print(total_tip_male )
print(total_tip_female )
```

```python
amount=x[:,1]#  total amount of male and female
sex=x[:,3]
tip=x[:,2]#avg tip of male and female
total_tip_male = np.mean(tip[sex == 0])
total_tip_female = np.mean(tip[sex == 1])
total_amount_male = np.sum(amount[sex == 0])+ np.sum(tip[sex == 0]) #amount spent by male
total_amount_female = np.sum(amount[sex == 1])+ np.sum(tip[sex == 1]) #amount spent by female
print(total_amount_male)
print(total_amount_female)

```

```python
lunch=x[:,6]#how many male and female are going for lunch
sex=x[:,3]
lunch_male = np.sum((sex == 0) & (lunch==0))
lunch_female = np.sum((sex == 1) & (lunch==0))
print('malecount for lunch:',lunch_male)
print('malecount for lunch:',lunch_female)
```

```python
dinner=x[:,6]#how many male and female are going for dinner
sex=x[:,3]
dinner_male = np.sum((sex == 0) & (dinner==1))
dinner_female = np.sum((sex == 1) & (dinner==1))
print('malecount for dinner: ',dinner_male)
print('malecount for dinner:',dinner_female)
```
