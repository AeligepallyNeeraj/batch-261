```python
Q1 = tdf['fare'].quantile(0.25)
Q3 = tdf['fare'].quantile(0.75)
IQR = Q3 - Q1
```

```python
IQR
```

```python
# Define upper and lower bounds for identifying outliers
lower_whisker = Q1 - 1.5 * IQR
upper_whisker = Q3 + 1.5 * IQR

# Identify outliers based on IQR
outliers_iqr = tdf[abs(tdf['fare'] < lower_whisker) | (tdf['fare'] > upper_whisker)]

# Display outliers
print("\nOutliers identified using IQR:")
print(outliers_iqr[['fare']])
```

```python
tdf['fare'].clip()
```

```python
sns.histplot(data=tdf,x=tdf['fare'],bins=20)
```

```python
lower_whisker = 0
upper_whisker = 65.6344

# Clip 'fare' values to the defined thresholds
tdf['fare_clipped'] = tdf['fare'].clip(lower=lower_whisker, upper=upper_whisker)
```

```python
tdf['fare_clipped']
```

```python
sns.histplot(data=tdf,x=tdf['fare_clipped'],bins=20)
```
