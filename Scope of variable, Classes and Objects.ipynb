# scope of a variable

* A variable is only available from inside the region it iscreated. This is called scope.
* The the scope of a variable refers to the region of the code where it can be accessed. Python has two types of variable scopes: global and local.
* 

## scope of a variable is classified into 4 types:
1.Built-in scope                                               

2.Global scope

3.Enclosing scope

4.Local scope


## Built-in scope of a variable

* In this the variables are pre-defined 

```python
x=5 # buil-in variable
def fooOuter():
    x=10
    def fooInner():
        x=15
        print(x)
    fooInner()
    print(x)
print(x)    
```

* In the above code we define x=5 so this is the built in scope of varaible

## Global scope of variable

* A variable created in the main body of the Python code is a global variableand belongs to the global scope.

* A variable created outside of a function is global and can be used by anyone


```python
x = 300

def myfunc():
    x=20# this is created in the main body of the function
    print(x)

myfunc()
```

```python
x=5 # buil-in variable
def fooOuter():
    x=10# global variable
    def fooInner():
        x=15
        print(x)
    fooInner()
    


 
```

```python
fooOuter()
```

## Enclosed scope

*  a python variable scope that isn’t global or local is nonlocal. This is also called the enclosing scope

* this scope of variable is with in the inner function but not in the inner function body 


    
## local scope of varaible

 

* Local variables in Python are those which are initialized inside a function and belong only to that particular function. It cannot be accessed anywhere outside the function. Let’s see how to create a local variable.

* a variable which is defined inside a function body is known as local variable

```python
x=5# global variable
def fooOuter():
    x=10# enclosed variable
    def fooInner():
        x=15#local variable
        print(x)
    fooInner()
    print(x)
```

```python
fooOuter()
```

## Non-Local scope of variable

* we use non-local variable to say that the value we get takes the outside function as inner function

```python
x=5
def fooOuter():
    x=10
    print(f'x was assigned a value{x} inside the outer function')
    print(f'value of x before entering inner func is{x}')
    
    def fooInner():
        #non local x
        print(f' x is a non- local var')
        x=15
        print(f'x was assigned a value{x} inside the inner function')
    fooInner()
    print(f'the value of x after exiting inner func is{x}')
```

```python
fooOuter()
```

```python
x=5
def fooOuter():
    x=10
    print(f'x was assigned a value{x} inside the outer function')
    print(f'value of x before entering inner func is{x}')
    
    def fooInner():
        nonlocal x
    
        x=15
        print(f'x was assigned a value{x} inside the inner function')
    fooInner()
    print(f'the value of x after exiting inner func is{x}')
```

```python
fooOuter()
```

## using global keyword 
* In this we get the values of local and enclosed as same by using the global varible

```python
x=5
def fooOuter():
    global x
    x=10
    print(f'x was assigned a value{x} inside the outer function')
    print(f'value of x before entering inner func is{x}')
    
    def fooInner():
        
        x=15
        print(f'x was assigned a value{x} inside the inner function')
    fooInner()
    print(f'the value of x after exiting inner func is{x}')
```
